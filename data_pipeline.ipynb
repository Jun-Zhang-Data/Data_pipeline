```python
import numpy as np

loans = []
for url in loan_urls:
    df = pd.read_csv(url, index_col=None, header=0)
    loans.append(df)

merged_loans = pd.concat(loans, axis=0, ignore_index=True)
```

##### Merge data for rows in webvisit dataframe that share the same webvisit_id

```python
df_visits['timestamp'] = df_visits.groupby(['webvisit_id'])['timestamp'].transform(lambda x : ','.join(x))
```

```python
df_visits['referrer'] = df_visits.groupby(['webvisit_id'])['referrer'].transform(lambda x : ','.join(x))
```

```python
df_visits['campaign_name'] = df_visits.groupby(['webvisit_id'])['campaign_name'].transform(lambda x : ','.join(x))
```

```python
df_visits.drop_duplicates(subset=['webvisit_id', 'timestamp', 'referrer', 'campaign_name'], keep='first', inplace=True)
```

```python
df_visits['webvisit_id'] = df_visits['webvisit_id'].astype(int)
```

```python
merged_loans = merged_loans.merge(df_visits, on='webvisit_id', how='left')
merged_loans.head(100)
```
